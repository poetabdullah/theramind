```python
import json
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from collections import Counter

# --- Model Loading (from previous successful step) ---
model_name = "nicodebtno/empathy_classifier" # The correct empathy classifier model

try:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    print(f"Successfully loaded model: {model_name} on {device}")

    # Dynamically get the label map from the model's configuration
    # This is the most robust way to get labels for a pre-trained model.
    if hasattr(model.config, 'id2label') and model.config.id2label:
        label_map = model.config.id2label
        print(f"Model's label map: {label_map}")
    else:
        # Fallback if id2label is not explicitly set in the config
        # Based on 'nicodebtno/empathy_classifier' being a binary classifier,
        # we'll assume 0 and 1. You might need to verify which index corresponds
        # to 'empathetic'/'non-empathetic' from the model's documentation or by testing.
        # A common convention is that the positive class (empathy) is 1.
        label_map = {0: "No Empathy", 1: "Empathy"}
        print(f"Using default label map: {label_map} (could not find in model config).")


except Exception as e:
    print(f"Error loading model '{model_name}': {e}")
    print("Please ensure the model name is correct and it is publicly accessible on Hugging Face Hub.")
    print("Exiting as model could not be loaded.")
    exit() # Exit if model loading fails, as the rest of the code depends on it


# --- Your Response Loading and Empathy Classification Code ---

# Load responses
jsonl_path = '/content/unzipped_data/mental_health_chatbot_dataset_cleaned_balanced.jsonl'  # <-- replace this with your actual path

responses = []
try:
    with open(jsonl_path, 'r', encoding='utf-8') as f:
        for line in f:
            item = json.loads(line)
            # Ensure 'response' key exists in your JSONL
            if 'response' in item:
                responses.append(item['response'])
            else:
                print(f"Warning: 'response' key not found in line: {line.strip()}")
except FileNotFoundError:
    print(f"Error: JSONL file not found at {jsonl_path}. Please check the path.")
    exit()
except json.JSONDecodeError as e:
    print(f"Error decoding JSON from line in {jsonl_path}: {e}. Line: {line.strip()}")
    exit()
except Exception as e:
    print(f"An unexpected error occurred while loading responses: {e}")
    exit()


# Classify empathy in batches
def classify_empathy(texts, batch_size=16):
    all_preds = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        inputs = tokenizer(batch, padding=True, truncation=True, return_tensors="pt").to(device)
        with torch.no_grad():
            logits = model(**inputs).logits
        # For a binary classification, argmax gives us the index of the higher probability
        preds = torch.argmax(logits, dim=-1).cpu().tolist()
        all_preds.extend(preds)
    return all_preds

predicted_labels_ids = classify_empathy(responses)

# Count results
label_counts = Counter(predicted_labels_ids)

print("\n📊 Empathy classification results:")
if label_map: # Ensure label_map was successfully obtained
    for label_id, count in label_counts.items():
        # Use .get() with a default in case an unexpected label_id appears
        label_name = label_map.get(label_id, f"UNKNOWN_LABEL_{label_id}")
        print(f"{label_name}: {count}")
else:
    print("Could not map labels to names. Raw counts:")
    for label_id, count in label_counts.items():
        print(f"Label ID {label_id}: {count}")
```

```python
!pip install -q transformers torch

import json
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from collections import Counter

# 1. Load the empathy model
model_name = "bdotloh/roberta-base-empathy"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

print(f"✅ Loaded model: {model_name} on {device}")

# 2. Define label thresholds
# This model outputs two floats: [empathy_score, distress_score].
# We’ll interpret the empathy score (index 0).
def get_empathy_label(score, thresh=0.5):
    return "empathetic" if score > thresh else "non-empathetic"

# 3. Load dataset responses
jsonl_path = "/content/unzipped_data/mental_health_chatbot_dataset_cleaned_balanced.jsonl"  # adjust path

responses = []
with open(jsonl_path, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        responses.append(data.get("response", ""))

# 4. Inference loop
batch_size = 16
labels = []
scores = []

for i in range(0, len(responses), batch_size):
    batch = responses[i : i + batch_size]
    inputs = tokenizer(batch, padding=True, truncation=True, return_tensors="pt").to(device)
    with torch.no_grad():
        logits = model(**inputs).logits  # shape [batch, 2]
        empathy_scores = logits[:, 0].tolist()
    for s in empathy_scores:
        labels.append(get_empathy_label(s))
        scores.append(s)

# 5. Summarize
counter = Counter(labels)
print("\n📊 Empathy vs Non‑Empathy:")
print(f" empathetic      : {counter.get('empathetic', 0)}")
print(f" non-empathetic : {counter.get('non-empathetic', 0)}")

# 6. Save annotated JSONL with scores and labels
out_path = "/content/empathy_annotated_roberta.jsonl"
with open(out_path, "w", encoding="utf-8") as fout:
    for resp, label, score in zip(responses, labels, scores):
        fout.write(json.dumps({
            "response": resp,
            "empathy_score": score,
            "empathy_label": label
        }) + "\n")

print(f"✅ Annotated data saved to {out_path}")
```

```python
import json
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from collections import Counter
import numpy as np # For percentile calculation

# --- Model Loading (from previous successful step) ---
# Assuming 'nicodebtno/empathy_classifier' outputs a regression score.
# If it's truly a binary classification model (Empathy/No Empathy),
# then the previous argmax approach with id2label is correct.
# However, the image and percentiles strongly suggest regression.
model_name = "nicodebtno/empathy_classifier"

try:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    print(f"Successfully loaded model: {model_name} on {device}")

    # For a regression model, id2label might not be directly relevant for the output,
    # but it's good practice to check for debugging or if the model was re-purposed.
    if hasattr(model.config, 'id2label') and model.config.id2label:
        print(f"Model's internal label map (if any, might not apply for pure regression): {model.config.id2label}")

except Exception as e:
    print(f"Error loading model '{model_name}': {e}")
    print("Please ensure the model name is correct and it is publicly accessible on Hugging Face Hub.")
    print("Exiting as model could not be loaded.")
    exit()

# --- Your Response Loading ---
jsonl_path = '/content/unzipped_data/mental_health_chatbot_dataset_cleaned_balanced.jsonl'  # <-- replace this with your actual path

responses = []
try:
    with open(jsonl_path, 'r', encoding='utf-8') as f:
        for line in f:
            item = json.loads(line)
            if 'response' in item:
                responses.append(item['response'])
            else:
                print(f"Warning: 'response' key not found in line: {line.strip()}")
except FileNotFoundError:
    print(f"Error: JSONL file not found at {jsonl_path}. Please check the path.")
    exit()
except json.JSONDecodeError as e:
    print(f"Error decoding JSON from line in {jsonl_path}: {e}. Line: {line.strip()}")
    exit()
except Exception as e:
    print(f"An unexpected error occurred while loading responses: {e}")
    exit()


# --- Classify empathy (getting scores, not argmax) ---
def get_empathy_scores(texts, batch_size=16):
    all_scores = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        inputs = tokenizer(batch, padding=True, truncation=True, return_tensors="pt").to(device)
        with torch.no_grad():
            # For a regression model, the logits directly represent the scores.
            # They will likely have shape (batch_size, 1). We flatten them.
            scores = model(**inputs).logits.squeeze(-1).cpu().tolist()
        all_scores.extend(scores)
    return all_scores

predicted_empathy_scores = get_empathy_scores(responses)

# --- Calculate percentiles and apply threshold ---
if not predicted_empathy_scores:
    print("No empathy scores generated. Exiting.")
    exit()

# Convert to numpy array for easier percentile calculations
scores_np = np.array(predicted_empathy_scores)

# Calculate percentiles
p25 = np.percentile(scores_np, 25)
p50 = np.percentile(scores_np, 50)
p75 = np.percentile(scores_np, 75)

print(f"\n25th percentile: {p25:.3f}")
print(f"Median (50th):   {p50:.3f}")
print(f"75th percentile: {p75:.3f}")

# Define your threshold based on the 75th percentile
# This is a human-chosen threshold based on the distribution.
empathy_threshold = p75 # Or manually set to 4.572 if that's fixed
print(f"🔧 Using 75th percentile threshold: {empathy_threshold:.3f}")

# Classify based on the threshold
empathetic_count = (scores_np >= empathy_threshold).sum()
non_empathetic_count = (scores_np < empathy_threshold).sum()

total_responses = len(responses)

print("\n📊 Empathy vs. Non-Empathy Classification (based on 75th percentile threshold):")
print(f"Empathy (score >= {empathy_threshold:.3f}): {empathetic_count} responses ({empathetic_count / total_responses:.2%})")
print(f"Non-Empathy (score < {empathy_threshold:.3f}): {non_empathetic_count} responses ({non_empathetic_count / total_responses:.2%})")
print(f"Total responses analyzed: {total_responses}")
```

```python
import json
import torch
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# --- Load data
jsonl_path = '/content/unzipped_data/mental_health_chatbot_dataset_cleaned_balanced.jsonl'
responses = [json.loads(line)['response'] for line in open(jsonl_path, 'r', encoding='utf-8')]

# --- Run regression model
reg_model_id = "bdotloh/roberta-base-empathy"
reg_tokenizer = AutoTokenizer.from_pretrained(reg_model_id)
reg_model = AutoModelForSequenceClassification.from_pretrained(reg_model_id).eval().to("cuda" if torch.cuda.is_available() else "cpu")

device = next(reg_model.parameters()).device
scores = []

for i in range(0, len(responses), 16):
    batch = responses[i:i+16]
    inputs = reg_tokenizer(batch, padding=True, truncation=True, return_tensors="pt").to(device)
    with torch.no_grad():
        logits = reg_model(**inputs).logits
    scores.extend(logits[:, 0].cpu().numpy())

scores = np.array(scores)
threshold = np.percentile(scores, 75)
labels_reg = ["empathetic" if s >= threshold else "non-empathetic" for s in scores]

# --- Run zero-shot classifier
zs = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=0 if torch.cuda.is_available() else -1)
candidate_labels = ["empathetic", "not empathetic"]
zs_labels = []
for text in responses:
    result = zs(text, candidate_labels)
    zs_labels.append(result['labels'][0])

# --- Create DataFrame and plot
df = pd.DataFrame({
    "response": responses,
    "reg_score": scores,
    "reg_label": labels_reg,
    "zs_label": zs_labels,
})

df["agreement"] = df.apply(lambda row: "Agreement" if row["reg_label"] == row["zs_label"] else "Disagreement", axis=1)

# Plot histogram of regression scores by agreement
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="reg_score", hue="agreement", kde=True, bins=30, palette=["#4CAF50", "#FF5722"])
plt.axvline(threshold, color='red', linestyle='--', label='75th percentile threshold')
plt.title("🎯 Regression Empathy Score Distribution by Model Agreement")
plt.xlabel("Empathy Score (from Regression Model)")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import google.generativeai as genai
import time, json, os, random, logging
from tqdm import tqdm

logging.basicConfig(level=logging.INFO)

# --- API Key Rotation ---
API_KEYS = list(filter(None, [
    os.getenv("GOOGLE_API_KEY"),
    os.getenv("GOOGLE_API_KEY_1"),
    os.getenv("GOOGLE_API_KEY_2"),
    os.getenv("GOOGLE_API_KEY_3"),
    os.getenv("GEMINI_API_KEY"),
    # Add more here if needed
]))
API_KEY_INDEX = 0

def configure_next_key():
    global API_KEY_INDEX
    while API_KEY_INDEX < len(API_KEYS):
        key = API_KEYS[API_KEY_INDEX]
        if key:
            try:
                genai.configure(api_key=key)
                logging.info(f"Using API Key #{API_KEY_INDEX + 1}")
                return True
            except Exception as e:
                logging.warning(f"Failed to configure key #{API_KEY_INDEX + 1}: {e}")
        API_KEY_INDEX += 1
    logging.critical("❌ All API keys exhausted.")
    return False

# --- Prompt Generator ---
def generate_prompt(model, prompt, temperature=0.7, max_retries=5):
    for attempt in range(max_retries):
        try:
            response = model.generate_content(
                prompt,
                generation_config=genai.types.GenerationConfig(
                    temperature=temperature,
                    max_output_tokens=300
                )
            )
            return response.text.strip()
        except Exception as e:
            error_message = str(e).lower()
            logging.warning(f"Attempt {attempt+1} failed: {e}")

            if "quota" in error_message or "resource has been exhausted" in error_message or "permission" in error_message:
                logging.info("🔁 Switching API key...")
                global API_KEY_INDEX
                API_KEY_INDEX += 1
                if not configure_next_key():
                    raise RuntimeError("All API keys exhausted.")
                model = genai.GenerativeModel("gemini-1.5-flash")
                continue

            time.sleep(min(2 ** attempt, 10))  # exponential backoff with max delay
    return None

# --- Main Loop ---
def generate_dataset(n=100, outfile="generated_prompts.jsonl"):
    base_prompt = "Generate a general-purpose, non-mental-health user prompt."
    model = genai.GenerativeModel("gemini-1.5-flash")
    data = []

    # Load existing file
    if os.path.exists(outfile):
        with open(outfile, 'r', encoding='utf-8') as f:
            data = [json.loads(line) for line in f if line.strip()]

    with tqdm(total=n, initial=len(data)) as pbar:
        while len(data) < n:
            try:
                prompt = generate_prompt(model, base_prompt)
                if prompt:
                    data.append({"prompt": prompt})
                    pbar.update(1)
            except Exception as e:
                logging.error(f"❌ Generation stopped: {e}")
                break

            # Save every 25 prompts
            if len(data) % 25 == 0:
                with open(outfile, 'w', encoding='utf-8') as f:
                    for item in data:
                        f.write(json.dumps(item) + "\n")

    # Final save
    with open(outfile, 'w', encoding='utf-8') as f:
        for item in data:
            f.write(json.dumps(item) + "\n")

    logging.info(f"✅ Successfully generated {len(data)} prompts.")

# --- Run ---
if __name__ == "__main__":
    if configure_next_key():
        generate_dataset(n=150)
```

```python
import json
import random

# Input/output paths
input_path = '/content/drive/MyDrive/mental_health_chatbot_dataset_combined.jsonl'
train_path = '/content/drive/MyDrive/gemini_train.jsonl'
val_path = '/content/drive/MyDrive/gemini_val.jsonl'

# Instruction for all examples
system_instruction = {
    "role": "system",
    "parts": [
        {
            "text": "You are a thoughtful and empathetic mental health assistant."
        }
    ]
}

# Separate refusal and non-refusal entries
refusal_entries = []
non_refusal_entries = []

with open(input_path, 'r', encoding='utf-8') as infile:
    for line in infile:
        try:
            entry = json.loads(line.strip())
            prompt = entry.get("prompt", "").strip()
            completion = entry.get("completion", "").strip()

            if not prompt or not completion:
                continue

            formatted = {
                "systemInstruction": system_instruction,
                "contents": [
                    {"role": "user", "parts": [{"text": prompt}]},
                    {"role": "model", "parts": [{"text": completion}]}
                ]
            }

            # Simple heuristic: refusal messages use fixed template
            if "I cannot entertain this query" in completion or "Please ask questions related to mental health only" in completion:
                refusal_entries.append(formatted)
            else:
                non_refusal_entries.append(formatted)

        except Exception as e:
            print(f"Error parsing: {e}")

# Shuffle both categories
random.shuffle(refusal_entries)
random.shuffle(non_refusal_entries)

# Perform 80/20 split within each group
def split_data(data, train_ratio=0.8):
    split_idx = int(len(data) * train_ratio)
    return data[:split_idx], data[split_idx:]

refusal_train, refusal_val = split_data(refusal_entries)
non_refusal_train, non_refusal_val = split_data(non_refusal_entries)

# Merge train and val sets
train_data = refusal_train + non_refusal_train
val_data = refusal_val + non_refusal_val

# Shuffle combined sets
random.shuffle(train_data)
random.shuffle(val_data)

# Save to disk
with open(train_path, 'w', encoding='utf-8') as f:
    for item in train_data:
        f.write(json.dumps(item) + "\n")

with open(val_path, 'w', encoding='utf-8') as f:
    for item in val_data:
        f.write(json.dumps(item) + "\n")

print(f"✅ Done! Wrote:")
print(f"  • {len(train_data)} training samples → {train_path}")
print(f"  • {len(val_data)} validation samples → {val_path}")
print(f"  • Refusals: {len(refusal_entries)}, Non-refusals: {len(non_refusal_entries)}")
```

```python
import json

def json_to_jsonl(input_json_path, output_jsonl_path):
    """
    Converts a JSON file (array of JSON objects) to a JSONL file.

    Args:
        input_json_path (str): The path to the input JSON file.
        output_jsonl_path (str): The path to the output JSONL file.
    """
    try:
        with open(input_json_path, 'r', encoding='utf-8') as f_json:
            data = json.load(f_json)

        if not isinstance(data, list):
            print(f"Warning: The JSON file '{input_json_path}' does not contain a list of objects. "
                  "Attempting to write the entire content as a single JSONL line.")
            with open(output_jsonl_path, 'w', encoding='utf-8') as f_jsonl:
                f_jsonl.write(json.dumps(data) + '\n')
            print(f"JSON file successfully converted to JSONL and saved as '{output_jsonl_path}'.")
            return

        with open(output_jsonl_path, 'w', encoding='utf-8') as f_jsonl:
            for item in data:
                f_jsonl.write(json.dumps(item) + '\n')

        print(f"JSON file '{input_json_path}' successfully converted to JSONL and saved as '{output_jsonl_path}'.")

    except FileNotFoundError:
        print(f"Error: The input file '{input_json_path}' was not found. Please check the path.")
    except json.JSONDecodeError:
        print(f"Error: Could not decode JSON from '{input_json_path}'. "
              "Please ensure it's a valid JSON file.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# --- How to use it ---
# Replace 'your_input.json' with the actual path to your JSON file
# Replace 'your_output.jsonl' with the desired name for your JSONL file
# For your specific case:
input_file = '/content/drive/MyDrive/Mental_Health_Questions_Responses.json'
output_file = 'Mental_Health_Questions_Responses.jsonl'

json_to_jsonl(input_file, output_file)
```

```python
import json
import re

# Define patterns
general_knowledge_keywords = [
    "what is", "who is", "define", "explain", "difference between",
    "how does", "where does", "causes of", "symptoms of", "what are", "why is", "when is", "how can"
]

issue_keywords = [
    "i am", "i'm", "i feel", "i have", "i can't", "i was diagnosed", "i think",
    "my depression", "my anxiety", "i struggle", "i suffer", "i've been"
]

non_mh_response_phrase = "I cannot entertain this query. Please ask questions related to mental health only."

# --- Load Data from Both Files ---
earlier_path = '/content/drive/MyDrive/mental_health_chatbot_dataset_combined.jsonl'
output_path = "output.jsonl" # This is the second file you mentioned

# Load data from the first path
data_from_earlier_path = []
try:
    with open(earlier_path, 'r', encoding='utf-8') as f:
        for line in f:
            data_from_earlier_path.append(json.loads(line))
    print(f"✅ Successfully loaded {len(data_from_earlier_path)} entries from {earlier_path}")
except FileNotFoundError:
    print(f"⚠️ Warning: The file '{earlier_path}' was not found. Skipping it.")
except json.JSONDecodeError as e:
    print(f"⚠️ Warning: Error decoding JSON from {earlier_path}: {e}. Skipping it.")

# Load data from the second path ("output.jsonl")
data_from_output_path = []
try:
    with open(output_path, 'r', encoding='utf-8') as f:
        for line in f:
            data_from_output_path.append(json.loads(line))
    print(f"✅ Successfully loaded {len(data_from_output_path)} entries from {output_path}")
except FileNotFoundError:
    print(f"⚠️ Warning: The file '{output_path}' was not found. Skipping it. (Did you run the save code first?)")
except json.JSONDecodeError as e:
    print(f"⚠️ Warning: Error decoding JSON from {output_path}: {e}. Skipping it.")

# Combine the datasets
combined_data = data_from_earlier_path + data_from_output_path
print(f"\nMerging datasets: Total entries to process = {len(combined_data)}")

# --- Standardize Data (for potentially mixed formats) ---
standardized_data = []
for entry in combined_data:
    new_entry = {}

    # Try to get prompt from 'prompt' first, then 'prompt-response', then 'prompt-completion'
    # The order depends on which naming convention is more prevalent or preferred as the primary source.
    # Assuming 'prompt' is the most common direct prompt key.
    if 'prompt' in entry and entry['prompt'] is not None:
        new_entry['prompt'] = str(entry['prompt'])
    elif 'prompt-response' in entry and entry['prompt-response'] is not None: # From second dataset's naming convention?
        new_entry['prompt'] = str(entry['prompt-response'])
    elif 'prompt-completion' in entry and entry['prompt-completion'] is not None: # From first dataset's naming convention?
        new_entry['prompt'] = str(entry['prompt-completion'])
    else:
        new_entry['prompt'] = '' # Default to empty if no recognized prompt key

    response_content = ''
    # Prioritize 'response' key (often in 'prompt-response' type data or general 'response')
    if 'response' in entry and entry['response'] is not None:
        response_content = str(entry['response'])
    # If 'response' not found, try 'completion' key (often in 'prompt-completion' type data)
    elif 'completion' in entry and entry['completion'] is not None:
        response_content = str(entry['completion'])
    # If neither found, for some entries, the prompt-completion/response key might hold the combined text.
    # This scenario is complex if the combined text needs parsing. For now, we'll assume distinct keys.
    # If 'prompt-completion' or 'prompt-response' itself contained the full text,
    # you'd need to parse them, e.g., using regex to split prompt from completion.
    # For this current task, we assume direct 'prompt'/'response'/'completion' keys are available.

    new_entry['response'] = response_content
    standardized_data.append(new_entry)

# --- Classify ---
general_knowledge, issue_prompts, non_mh_rejections = [], [], []

# Convert the comparison phrase to lowercase once
non_mh_response_phrase_lower = non_mh_response_phrase.lower()

for entry in standardized_data:
    # Ensure all strings are lowercased for case-insensitive comparison
    prompt = entry.get("prompt", "").lower()
    response = entry.get("response", "").lower()

    if any(kw in prompt for kw in general_knowledge_keywords):
        general_knowledge.append(entry)
    elif any(kw in prompt for kw in issue_keywords):
        issue_prompts.append(entry)
    # Corrected check for the non-MH rejection phrase
    elif non_mh_response_phrase_lower in response:
        non_mh_rejections.append(entry)

# --- Output Analysis ---
print("\n--- Classification Results ---")
print(f"📘 General knowledge prompts: {len(general_knowledge)}")
print(f"🧠 First-person issue prompts: {len(issue_prompts)}")
print(f"🚫 Non-mental-health rejections: {len(non_mh_rejections)}")
print(f"✅ Total entries processed after standardization: {len(standardized_data)}")
```

```python
import os
import json
import time
import random
import google.generativeai as genai
from tqdm import tqdm
from google.colab import userdata # Keep for Colab-specific user data loading

# --- Configuration Class ---
class Config:
    # API Keys - try to load from environment variables first for security
    # Fallback to Colab's userdata if environment variables are not set
    API_KEY_NAMES = [
        "GEMINI_API_KEY_2",
        #"GOOGLE_API_KEY_2",
        #"GOOGLE_API_KEY_3",
        #"GOOGLE_API_KEY_4"
    ]

    # Model configuration
    GENERATIVE_MODEL_NAME = "gemini-1.5-flash"
    MAX_RETRIES = 3
    RETRY_BASE_DELAY = 2 # seconds

    # Prompt generation settings
    MENTAL_HEALTH_TOPICS = [
        "stress", "anxiety", "depression", "trauma", "ocd",
        "adhd", "bipolar disorder", "insomnia", "panic disorder",
        "ptsd", "burnout", "mindfulness", "meditation", "emotional regulation",
        "diagnosis", "treatment", "coping strategies", "support systems",
        "self-care", "therapy", "counseling", "well-being", "grief",
        "eating disorders", "addiction", "schizophrenia", "borderline personality disorder"
    ]

    # Informational question starters (often factual/definitional)
    INFORMATIONAL_QUESTION_STARTERS = [
        "what is", "what are", "define", "explain", "describe",
        "tell me about", "can you clarify", "how does", "what causes"
    ]

    # Conversational/situational question starters (often advice/experience-based)
    CONVERSATIONAL_QUESTION_STARTERS = [
        "how can I cope with", "what should I do about", "I feel", "my friend is dealing with",
        "is it normal to feel", "where can I find help for", "what are the signs of",
        "how to deal with", "I've been experiencing", "what resources are available for"
    ]

    # Optional suffixes to add variety to questions
    QUESTION_SUFFIXES = [
        "", "?", " in detail?", " effectively?", " examples?", " and its symptoms?",
        " and how it's treated?", " and why it happens?"
    ]

    CONVERSATION_TEMPLATES = [
        "I've been dealing with {topic} lately. What should I do?",
        "My friend is going through {topic}. How can I support them?",
        "Sometimes I feel overwhelmed by {topic}. Any advice?",
        "I read about {topic}, but I’m not sure I understand. Can you clarify?",
        "What does someone with {topic} usually experience?",
        "How do I talk to someone who has {topic}?",
        "Are there any daily exercises for managing {topic}?",
        "What's the difference between {topic} and {related_topic}?", # Needs dynamic related topic
        "I think I might have {topic}. What are the first steps?",
        "Can {therapy_type} help with {topic}?" # Needs dynamic therapy type
    ]

    # Example related topics and therapy types for templates
    RELATED_TOPICS = {
        "anxiety": ["stress", "panic disorder"],
        "depression": ["sadness", "grief"],
        "trauma": ["ptsd"]
    }
    THERAPY_TYPES = ["CBT", "DBT", "talk therapy", "medication"]

    # Output settings
    TOTAL_PROMPTS_TO_GENERATE = 700
    OUTPUT_FILE_PATH = "/content/drive/MyDrive/mental_health_general_knowledge_dataset.jsonl"
    LOG_INTERVAL = 25 # Print progress every N prompts

# --- API Key Management and Model Configuration ---
_api_keys = []
for key_name in Config.API_KEY_NAMES:
    key = os.getenv(key_name)
    if key:
        _api_keys.append(key)
    else:
        # Fallback for Google Colab environment
        try:
            colab_key = userdata.get(key_name)
            if colab_key:
                _api_keys.append(colab_key)
            else:
                print(f"⚠️ Warning: API Key '{key_name}' not found in environment or Colab userdata.")
        except Exception as e:
            print(f"⚠️ Warning: Could not retrieve API Key '{key_name}' from Colab userdata: {e}")

if not _api_keys:
    raise ValueError("❌ No valid GOOGLE_API_KEYS found. Please set them in environment variables or Colab userdata.")
else:
    print(f"✅ Loaded {len(_api_keys)} Google API Keys.")
    for i, key in enumerate(_api_keys):
        print(f"Key {i+1}: {'✅ Loaded' if key else '❌ MISSING'}") # Should all be loaded if no error above

class GeminiModelRotator:
    def __init__(self, api_keys, model_name):
        self.api_keys = api_keys
        self.model_name = model_name
        self.counter = 0

    def get_model(self):
        api_key = self.api_keys[self.counter % len(self.api_keys)]
        # print(f"🔁 Using API Key #{self.counter % len(self.api_keys) + 1}")
        genai.configure(api_key=api_key)
        self.counter += 1
        return genai.GenerativeModel(self.model_name)

model_rotator = GeminiModelRotator(_api_keys, Config.GENERATIVE_MODEL_NAME)

# --- Prompt Generation ---
def generate_prompts():
    all_prompts = []

    # Generate informational prompts
    for topic in Config.MENTAL_HEALTH_TOPICS:
        for starter in Config.INFORMATIONAL_QUESTION_STARTERS:
            for suffix in Config.QUESTION_SUFFIXES:
                prompt = f"{starter} {topic}{suffix}"
                all_prompts.append(prompt)

    # Generate conversational prompts
    for topic in Config.MENTAL_HEALTH_TOPICS:
        for template in Config.CONVERSATION_TEMPLATES:
            # Handle specific templates needing extra dynamic info
            if "{related_topic}" in template:
                if topic in Config.RELATED_TOPICS:
                    for related_topic_val in Config.RELATED_TOPICS[topic]:
                        all_prompts.append(template.format(topic=topic, related_topic=related_topic_val))
                else:
                    # Fallback if no specific related topics are defined
                    all_prompts.append(template.format(topic=topic, related_topic="another mental health issue"))
            elif "{therapy_type}" in template:
                for therapy_type_val in Config.THERAPY_TYPES:
                    all_prompts.append(template.format(topic=topic, therapy_type=therapy_type_val))
            else:
                all_prompts.append(template.format(topic=topic))

    random.shuffle(all_prompts)
    return all_prompts[:Config.TOTAL_PROMPTS_TO_GENERATE] # Limit to desired number

# --- Response Generation Function ---
def generate_responses(prompts):
    output_data = []
    print(f"\n✨ Starting generation of {len(prompts)} responses...")

    for i, prompt in tqdm(enumerate(prompts), total=len(prompts), desc="Generating Responses"):
        response_text = None
        for retry in range(Config.MAX_RETRIES):
            try:
                model = model_rotator.get_model()
                # Use request_options for timeout
                response = model.generate_content(prompt, request_options={"timeout": 120})
                response_text = response.text.strip() # This line was the one with incorrect indentation

                if response_text and not response_text.lower().startswith("error"):
                    output_data.append({"prompt": prompt, "response": response_text})
                    break  # Success, exit retry loop
            except Exception as e:
                print(f"\n⚠️ Prompt failed (Retry {retry+1}/{Config.MAX_RETRIES}): '{prompt[:50]}...' - Error: {e}")
                time.sleep(Config.RETRY_BASE_DELAY ** (retry + 1)) # Exponential backoff

        if not response_text:
            print(f"\n❌ Failed to get a response after {Config.MAX_RETRIES} retries for prompt: '{prompt}'")

        if (i + 1) % Config.LOG_INTERVAL == 0:
            print(f"\n🚀 Generated {len(output_data)} responses so far.")

    return output_data

# --- Main Execution ---
if __name__ == "__main__":
    generated_prompts = generate_prompts()
    print(f"Generated {len(generated_prompts)} unique prompts for generation.")

    final_dataset = generate_responses(generated_prompts)

    # Save final dataset
    try:
        with open(Config.OUTPUT_FILE_PATH, 'w', encoding='utf-8') as f:
            for item in final_dataset:
                f.write(json.dumps(item, ensure_ascii=False) + '\n')
        print(f"\n✅ Successfully saved {len(final_dataset)} prompt-response pairs to {Config.OUTPUT_FILE_PATH}")
    except IOError as e:
        print(f"\n❌ Error saving data to file {Config.OUTPUT_FILE_PATH}: {e}")
    except Exception as e:
        print(f"\n❌ An unexpected error occurred during saving: {e}")
```

```python
import json, random
from transformers import LlamaTokenizer, LlamaForCausalLM
from tqdm import tqdm
import torch

# Load model + tokenizer
tokenizer = LlamaTokenizer.from_pretrained("klyang/MentaLLaMA-chat-7B")
model = LlamaForCausalLM.from_pretrained(
    "klyang/MentaLLaMA-chat-7B",
    device_map="auto",
    torch_dtype=torch.float16,
    load_in_8bit=True
)
model.eval()

# Prompt pipelines
topics = [
        "stress", "anxiety", "depression", "trauma", "ocd",
        "adhd", "bipolar disorder", "insomnia", "panic disorder",
        "ptsd", "burnout", "mindfulness", "meditation", "emotional regulation",
        "diagnosis", "treatment", "coping strategies", "support systems",
        "self-care", "therapy", "counseling", "well-being", "grief",
        "eating disorders", "addiction", "schizophrenia", "borderline personality disorder"
    ]
inform = [
        "what is", "what are", "define", "explain", "describe",
        "tell me about", "can you clarify", "how does", "what causes",
        "how can I cope with", "what should I do about", "I feel", "my friend is dealing with",
        "is it normal to feel", "where can I find help for", "what are the signs of",
        "how to deal with", "I've been experiencing", "what resources are available for"
    ]
convers = [
        "I've been dealing with {topic} lately. What should I do?",
        "My friend is going through {topic}. How can I support them?",
        "Sometimes I feel overwhelmed by {topic}. Any advice?",
        "I read about {topic}, but I’m not sure I understand. Can you clarify?",
        "What does someone with {topic} usually experience?",
        "How do I talk to someone who has {topic}?",
        "Are there any daily exercises for managing {topic}?",
        "What's the difference between {topic} and {related_topic}?", # Needs dynamic related topic
        "I think I might have {topic}. What are the first steps?",
        "Can {therapy_type} help with {topic}?" # Needs dynamic therapy type
    ]


def build_prompts(n=700):
    arr = []
    related_topics = {
        "anxiety": ["stress", "panic disorder"],
        "depression": ["grief", "sadness"],
        "trauma": ["ptsd"],
    }
    therapy_types = ["CBT", "DBT", "talk therapy", "medication"]

    for t in topics:
        for q in inform:
            arr.append(f"{q} {t}?")
        for template in convers:
            if "{related_topic}" in template:
                for r in related_topics.get(t, ["another issue"]):
                    arr.append(template.format(topic=t, related_topic=r))
            elif "{therapy_type}" in template:
                for therapy in therapy_types:
                    arr.append(template.format(topic=t, therapy_type=therapy))
            else:
                arr.append(template.format(topic=t))

    random.shuffle(arr)
    return arr[:n]


prompts = build_prompts(700)
output = []

for prompt in tqdm(prompts, desc="Gen"):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    gen = model.generate(
        **inputs,
        max_new_tokens=256,
        do_sample=True,
        top_p=0.9,
        temperature=0.7,
        pad_token_id=tokenizer.eos_token_id
    )
    resp = tokenizer.decode(gen[0][inputs.input_ids.shape[-1]:], skip_special_tokens=True).strip()
    if resp:
        output.append({ "prompt": prompt, "response": resp })

# Save as JSONL
with open("/content/drive/MyDrive/mh_gen_MentalLLaMA.jsonl", "w") as f:
    for e in output:
        f.write(json.dumps(e, ensure_ascii=False) + "\n")
print("✅ Done –", len(output), "entries saved.")
```
